```python
#What percent of the total number of passengers on the Titanic belonged to each age group
df = pd.read_csv('/content/drive/MyDrive/AI/Titanic-Dataset.csv')
df.head()
```

```python
def classify_age(age):
    if age < 18:
        return "child"
    elif 18 <= age < 65:
        return "adult"
    else:
        return "senior"
df["age_group"] = df["Age"].apply(classify_age)
```

```python
total = len(df)
total
```

```python
group_counts = df["age_group"].value_counts()
group_percent = (group_counts / total) * 100
print("Percentage of Passengers by Age Group:\n")
print(group_percent)
```
